#Preprocesamiento

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler, OrdinalEncoder
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split

from sklearn import set_config
```

```python
y=df["Anemia"]
X=df.drop(columns=["Anemia"])
```

```python
df
```

```python
df.info()
```

```python
escalables = ['Births in last five years', "Age of respondent at 1st birth"]
nominales = ['Type of place of residence', 'Age in 5-year groups',"Have mosquito bed net for sleeping (from household questionnaire)"]
ordinales = ['Highest educational level', 'Wealth index combined', 'Smokes cigarettes', 'Current marital status', 'Currently residing with husband/partner', 'Had fever in last two weeks', 'Taking iron pills, sprinkles or syrup']
```

```python
nom_selector = make_column_selector(dtype_include='object')
num_selector = make_column_selector(dtype_include='number')
ord_selector = make_column_selector(dtype_include='object')


# Transformadores
one_hot_encoder = OneHotEncoder()
standard_scaler = StandardScaler()
ordinal_encoder = OrdinalEncoder()

imputer_num = SimpleImputer(strategy="mean")
imputer_nom = SimpleImputer(strategy="most_frequent")
imputer_ord = SimpleImputer(strategy="most_frequent")
```

```python
col_transformer = make_column_transformer((make_pipeline(standard_scaler, imputer_num), escalables),
                                             (make_pipeline(one_hot_encoder, imputer_nom), nominales),
                                             (make_pipeline(ordinal_encoder, imputer_ord), ordinales),
                                             remainder = 'passthrough')

col_transformer
```

```python
col_transformer.fit(X)
```

```python
aux=col_transformer.transform(X)
X_procesado=pd.DataFrame(aux,columns=col_transformer.get_feature_names_out())
X_procesado.info()
X_procesado
```

Los resultados del %%time obtenemos que si aplicamos la reduccion de dimensionalidad, podemos obtener un menor tiempo de prediccion de nuestro modelo. Conclusion: Se tomara el modelo con PCA para reducir el tiempo de prediccion de nuestro modelo.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Escalamiento
scaler = StandardScaler()
X_train1 = scaler.fit_transform(X_train1)
X_test1 = scaler.transform(X_test1)

# Input shape
input_shape = X_train1.shape[1]
print(input_shape)
```

### Modelo con PCA 1

```python
model = Sequential()
model.add(Dense(9,
                input_dim = input_shape,
                activation = 'relu'))
model.add(Dense(5,
                activation = 'relu'))
model.add(Dense(1, activation = "sigmoid"))
model.summary()
```

```python
model.compile(loss = 'bce', optimizer = 'adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train1, y_train1,
                    validation_data = (X_test1, y_test1),
                    epochs=10)
```

```python
plt.plot(history.history['loss'], label='Train loss', color = "blue")
plt.plot(history.history['val_loss'], label='Validation Loss', color="red")
plt.title("PÃ©rdida (loss) durante el entrenamiento y la validaciÃ³n del modelo")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend();
```

```python
# Visualizar el accuracy
plt.plot(history.history['accuracy'], label='Train Accuracy', color = "blue")
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', color = "red")
plt.title("PrecisiÃ³n (accuracy) durante el entrenamiento y la validaciÃ³n del modelo")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend();
```

```python
loss, acc = model.evaluate(X_test1, y_test1,
                            batch_size=32)
print("Loss:", loss) # FunciÃ³n de pÃ©rdida
print("Accuracy:", acc)
```
